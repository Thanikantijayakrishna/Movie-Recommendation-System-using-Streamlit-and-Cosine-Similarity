```python
import streamlit as st
import joblib
import pandas as pd

# Load the saved models and data
tfidf_matrix = joblib.load('tfidf_matrix.pkl')
cosine_sim = joblib.load('cosine_sim.pkl')
indices = joblib.load('indices.pkl')
df = joblib.load('movie_df.pkl')

# Define the recommendation function
def recommend_movies(title):
    title = title.lower()

    if title not in indices:
        return ["Movie not found."]

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Top 10, exclude the searched movie
    movie_indices = [i[0] for i in sim_scores]

    return df['original_title'].iloc[movie_indices].tolist()

# ----------------- Streamlit UI ------------------
st.set_page_config(page_title="Movie Recommender", layout="centered")
st.title("🎬 Movie Recommendation System")

# Input box
movie_name = st.text_input("Enter a movie name:", placeholder="e.g., The Godfather")

# Button to trigger recommendation
if st.button("Get Recommendations"):
    if movie_name.strip() == "":
        st.warning("Please enter a movie name.")
    else:
        recommendations = recommend_movies(movie_name)
        if recommendations[0] == "Movie not found.":
            st.error("❌ Movie not found in the database.")
        else:
            st.success("✅ Recommended Movies:")
            for i, rec in enumerate(recommendations, 1):
                st.write(f"{i}. {rec}")
```
